```python
import pandas as pd
import pickle
from pycaret.regression import *
from sklearn.model_selection import train_test_split
from sklearn.ensemble import GradientBoostingRegressor
```

```python
data=pd.read_csv("data.csv")
```

```python
data.drop(['sl_no','ssc_b','hsc_b','mba_p'],axis=1,inplace=True)
data=pd.get_dummies(data,columns=['gender','hsc_s','degree_t','workex','specialisation','status'],drop_first=True)
data['sal']=data['salary']
data.drop(['salary'],inplace=True,axis=1)
data.fillna(0,inplace=True)
```

```python
X=data.iloc[:, 0:12]
y=data.iloc[:,12:]
X_train, X_test, y_train, y_test=train_test_split(X,y,train_size=0.75)
```

```python
model=GradientBoostingRegressor().fit(X_train,y_train)
pred=model.predict(X_test)
file=open('l.pkl','wb')
pickle.dump(model,file)
```
